```python
import pandas as pd
from scipy.stats import f # f-распределение и критические значения
import numpy as np
from statsmodels.formula.api import ols

# импорт данных
df = pd.read_csv('sleep75.csv') #df - название датасета

#print(df['sleep'])
#print(df[['sleep', 'male', 'marr']])
```

```python
# Задание1: Оцените регрессию и проверьте её значимость используя неробастный тест.
# 'sleep ~ male + marr + union + yngkid'
# Укажите тестовую статистику, критическое значение и сделайте вывод.
# Ответ округлите до 3 десятичных знаков. Уровень значимости 1%.
```

```python

# оцените регрессию - построить модель
mod1 = ols(formula='sleep ~ male + marr + union + yngkid', data=df)
res1 = mod1.fit() #  название модели
#print(res1.summary()) #распечатать полное summary
#res1.params #параметры модели - коэффициенты beta
```

оцените значимость - либо смотрим модель в целом либо отдельные коэффициенты (задано в условии задачи)
модель в целом : F-test (тест Фишера)
выбранный коэффициент: t-test (тест Фишера)

Робастный /не робастный - по каким формулам считаем.
робастный - более точный

проверка модели в целом - F тест, совместная значимость https://github.com/artamonoff/econometrica/blob/main/econometrica-1/jupyter-notebooks/f-test-joint.ipynb

```python
#F-тест (неробастный)
F_test = res1.f_test('male=marr=union=yngkid=0')
print(F_test) #F - тестовая статистика, p - p-value
```

```python
# Тестовая статистика и её P-значение
F_test.statistic, F_test.pvalue
```

```python
# уровень значимости - берем из условия
sign_level = 0.01
# Критическое значение F-распределения
f.ppf(q=1-sign_level, dfn=F_test.df_num, dfd=F_test.df_denom)
```

```python
# ответ на вопрос теста
# тестовая статистика: 0.948
# критическое значение: 3.346
# регрессия: не значима
```

вывод - если F test > F crit -> модель значима
если F test  < F crit -> модель не значима

```python
#F-тест (робастный)
# пересчитываем модель - по-другому считаем ошибки
res2 = mod1.fit(cov_type='HC3')

F_test = res2.f_test('male=marr=union=yngkid=0')
print(F_test)


```

```python
# уровень значимости -    аналогично неробастному тесту
sign_level = 0.01
# Критическое значение F-распределения
f.ppf(q=1-sign_level, dfn=F_test.df_num, dfd=F_test.df_denom)
```

```python
# Задание2: Оцените регрессию и проверьте её значимость используя неробастный тест.
# sleep ~ age + I(age ** 2) + smsa + yngkid + marг
# Укажите тестовую статистику, критическое значение и сделайте вывод.
# Ответ округлите до 3 десятичных знаков. Уровень значимости 5%.

# решение
# датасет, загрузили его в df

# было: sleep ~ male + marr + union + yngkid
# стало: sleep ~ age + I(age ** 2) + smsa + yngkid + marг
```

```python
import pandas as pd
from scipy.stats import f # f-распределение и критические значения
import numpy as np
from statsmodels.formula.api import ols

# импорт данных
df = pd.read_csv('sleep75.csv') #df - название датасета
```

```python
# оцените регрессию - построить модель
mod1 = ols(formula='sleep ~ age + age ** 2 + smsa + yngkid + marr', data=df)
res1 = mod1.fit() #  название модели
```

```python
#F-тест (неробастный)
F_test = res1.f_test('age=smsa=yngkid=marr=0')
print(F_test) #F - тестовая статистика, p - p-value
```

```python
# уровень значимости - берем из условия
sign_level = 0.05
# Критическое значение F-распределения
f.ppf(q=1-sign_level, dfn=F_test.df_num, dfd=F_test.df_denom)
```

```python
# ответ на вопрос теста
# тестовая статистика: 2.703
# критическое значение: 2.385
# регрессия: значима
```
